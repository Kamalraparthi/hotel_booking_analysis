## ==> 3. Data Cleaning and Preparation
---

* **Check shape and preview first few rows**

```python
print("shape of dataset: ", df.shape)
```

```python
df.head(10)
```

* **Dataset Info: Column Names & Data Types**

```python
df.info()
```

* **Statistical Summary of numerical columns**

```python
df.describe()
```

* **Check for Missing Values**

```python
df.isnull().sum().sort_values(ascending=False)
```

* **Handle missing values**

```python
df['children'] = df['children'].fillna(0)
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
df['country']= df['country'].fillna('Unknown')
```

```python
df[['children', 'agent', 'company', 'country']].isnull().sum()
```

* **Convert Date Columns to datetime type**

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df['reservation_status_date'].head()
```

* **Add New Feature: Total Stay Duration**

```python
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
import seaborn as sns
sns.histplot(df['total_stay'], bins=20)
```

* **Check for Duplicates**

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

df = df.drop_duplicates()
```

```python
df.shape
```

* **Check Unique Values in Key Columns**

```python
# Unique hotel types
print("Hotel types:", df['hotel'].unique())

# Unique meal types
print("Meal types:", df['meal'].unique())

# Unique market segments
print("Market segments:", df['market_segment'].unique())
```

* **Add Booking Status Text Column**

```python
df.loc[:, 'booking_status'] = df['is_canceled'].apply(lambda x: 'Canceled' if x == 1 else 'Not Canceled')
```

```python
df[['is_canceled', 'booking_status']].head()
```

* **Create a Full Arrival Date Column**

```python
df.loc[:, 'arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str)
)
```

```python
df[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 'arrival_date']].head()
```

* **Re-check Data Types and Missing Values**

```python
df.info()
df.isnull().sum()
```

## ==> 4. Exploratory Data Analysis (EDA) and Insights
---

* **Booking Counts by Hotel Type**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='hotel', hue='hotel', palette='Set2')
plt.title("Booking Counts by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")
plt.show()
```

**Insight:**
- City hotels have significantly more bookings compared to resort hotels.
- This suggests that guests prefer city hotels, possibly due to better location accessibility or business travel demand.

* **Booking Status (Canceled vs Not Canceled)**

```python
sns.countplot(data=df, x='booking_status', hue='booking_status', palette='coolwarm')
plt.title("Booking Cancellation Status")
plt.xlabel("Status")
plt.ylabel("Number of Bookings")
plt.show()
```

**Insight:**
- A large proportion of bookings have been canceled.
- High cancellation rates could indicate strict policies, pricing issues, or overbooking practices.

* **Monthly Booking Trends**

```python
# Ensure months are ordered
months_order = ['January','February','March','April','May','June',
                'July','August','September','October','November','December']

monthly_counts = df['arrival_date_month'].value_counts().reindex(months_order)

monthly_counts.plot(kind='bar', color='skyblue')
plt.title("Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.show()
```

**Insight:**
- Bookings are highest in **July and August**, likely due to summer vacations.
- Lowest bookings occur in **December and January**, indicating the off-season.
- This trend highlights the impact of seasonality on hotel demand.

* **Top 10 Countries by Bookings**

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values, color='teal')
plt.title("Top 10 Countries by Guest Count")
plt.xlabel("Country")
plt.ylabel("Number of Guests")
plt.show()
```

**Insight:**
- The majority of guests are from **Portugal, the UK, France, and Spain**.
- Portugal has the highest guest volume, suggesting strong domestic tourism.

* **Average Daily Rate (ADR) by Hotel**

```python
sns.boxplot(data=df, x='hotel', y='adr', hue='hotel', palette='pastel')
plt.title("ADR (Average Daily Rate) by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("ADR")
plt.ylim(0, 500)  # limit extreme outliers
plt.show()
```

**Insight:**
- City hotels tend to have a **wider spread and higher ADRs** compared to resort hotels.
- There are more price outliers in city hotels, indicating dynamic or premium pricing strategies.

* **Booking Cancellations by Hotel Type**

```python
sns.countplot(data=df, x='hotel', hue='booking_status', palette='Set1')
plt.title("Booking Status by Hotel Type")
plt.xlabel("Hotel")
plt.ylabel("Count")
plt.legend(title="Status")
plt.show()
```

**Insight:**
- **City hotels** show higher cancellation rates than resort hotels.
- This might be due to higher booking flexibility or OTA (online travel agent) usage in city properties.

* **Average Stay Duration by Hotel Type**

```python
avg_stay = df.groupby('hotel')['total_stay'].mean().reset_index()

sns.barplot(data=avg_stay, x='hotel', y='total_stay', hue='hotel', palette='spring')
plt.title("Average Stay Duration by Hotel Type")
plt.ylabel("Average Total Stay (Nights)")
plt.show()
```

**Insight:**
- Guests at **resort hotels** tend to stay longer on average than those at city hotels.
- This aligns with the nature of resort stays being more leisure/vacation-oriented.


* **Distribution of ADR (Average Daily Rate)**

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['adr'], bins=100, kde=True, color='teal')
plt.title("Distribution of Average Daily Rate (ADR)")
plt.xlabel("ADR")
plt.xlim(0, 500)
plt.show()
```

**Insight:**
- Most ADR values fall between **0 and 200**.
- There are noticeable outliers with very high ADRs â€” likely due to longer stays, luxury rooms, or group bookings.


* **Relationship Between Total Stay and ADR**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='total_stay', y='adr', hue='hotel', alpha=0.6)
plt.title("Total Stay vs ADR")
plt.xlabel("Total Stay (Nights)")
plt.ylabel("Average Daily Rate")
plt.ylim(0, 500)
plt.show()
```

**Insight:**
- ADR doesn't strongly increase with total stay â€” suggesting **per-night pricing is consistent**.
- Some shorter stays still show high ADRs â€” likely due to last-minute bookings or peak periods.


* **Correlation Heatmap (Numerical Features)**

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

**Insight:**
- `total_stay` is positively correlated with `adr`, which is expected.
- `is_canceled` has weak correlations with most features, implying that cancellation behavior is complex and may depend on categorical or external factors.

## ==> 5. Final Conclusion and Recommendations
---

### Summary of Findings:

- **City hotels** received more bookings but also had significantly **higher cancellation rates**.
- Most bookings occurred in **July and August**, indicating strong **seasonal demand**.
- **Portugal, the UK, and France** were the top guest-origin countries.
- Most bookings came via **Online Travel Agents (OTAs)**, while direct bookings were fewer.
- Guests typically stayed for **1 to 4 nights**, especially in city hotels.
- The **Average Daily Rate (ADR)** was higher for city hotels and during peak seasons.

---

### Business Recommendations:

1. **Reduce Cancellation Rate**:
   - Review cancellation policies, especially for city hotels.
   - Offer incentives for non-refundable bookings (e.g., discounts or upgrades).

2. **Boost Off-Season Demand**:
   - Launch marketing campaigns and discounts from **December to February**.
   - Offer bundled deals for longer stays or early bookings.

3. **Leverage Guest Demographics**:
   - Target top guest countries (e.g., Portugal, UK, France) with personalized marketing.
   - Improve multilingual support and country-specific offers.

4. **Increase Direct Bookings**:
   - Encourage direct bookings through website discounts or free add-ons (Wi-Fi, breakfast).
   - Promote loyalty programs for repeat customers.

5. **Dynamic Pricing Strategy**:
   - Adjust prices dynamically based on **stay duration**, **lead time**, and **season**.
   - Use ADR trends to optimize revenue management.

---

### Conclusion:

This analysis gives a comprehensive overview of booking behavior, guest demographics, and pricing in the hospitality domain. These insights can support data-driven decisions to **improve occupancy, reduce cancellations, and boost profitability**.
